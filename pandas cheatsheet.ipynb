### Changing data

```python
# # adding new column (also called Series)
# df['Total'] = df['HP'] + df['Attack'] + df['Defense']
# df.head() 

# another way to do the same thing as above
df['Total'] = df.iloc[:, 4:10].sum(axis = 1)
df.head()
```

```python
# drop a column (Series)
df = df.drop(columns = ['Total'])
df.head()
```

```python
#re-arrange the columns
cols = list(df.columns.values) # get the list of columns first
df = df[cols[0:4] + [cols[-1]] + cols[4:12]] # concatenate the lists accross columns
df.head()
```

### Saving our data 

```python
# save to csv with index
df.to_csv('modified.csv')
# save to csv without index
df.to_csv('modified.csv', index = False)

# save to excel with index
df.to_excel('modified.xlsx')
# save to excel without index
df.to_excel('modified.xlsx', index = False)

# save to txt with index
df.to_csv('modified.txt', sep = '\t')
# save to txt without index
df.to_csv('modified.txt', index = False, sep = '\t')
```

### Filtering

```python
# using df.loc to filter entries based on conditions
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')] # & sign and operation in pandas
df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')] # | sign for or operation in pandas
```

```python
# reset index on the resulting dataframes after performing operations
new_df = df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')]
new_df.reset_index() # saves old index as a column
new_df.reset_index(drop=True) # drops indexes truly
new_df.reset_index(drop=True, inplace = True) # drops old index and resets in place
```

```python
# using string functions
df.loc[df['Name'].str.contains('Mega')] # entries that contain 'Mega' in the Name series
df.loc[~df['Name'].str.contains('Mega')] # entries that do not contain 'Mega' in the Name series
```

```python
# contains method also accepts regex
import re
df.loc[df['Type 1'].str.contains('fire|grass', flags = re.I, regex = True)] # retrieve entries that have values fire or grass for 'Type 1'. re.I -> ignore match case sensitivity
```

### Conditional Changes

```python
df.loc[df['Type 1'] == 'Fire', 'Legendary'] = True # change the values of 'Legendary' series to true based on the 'Type 1' series
df
```

### Functions

```python
# df.groupby
df.groupby(['Type 1']).mean() # group by 'Type 1' and get the mean of each 'Type 1'
df.groupby(['Type 1']).sum() # group by 'Type 1' and get the sum of each 'Type 1'
df.groupby(['Type 1']).count() # group by 'Type 1' and get the count of each 'Type 1'
df.groupby(['Type 1', 'Type 2']).count() # group by multiple columns
```

### When there is lots of data

```python
# use 'chunksize' property to import dataset in chunks
new_df = pd.DataFrame(columns = df.columns)
for df in pd.read_csv('pokemon_data.csv', chunksize = 100): # read 100 rows each time
    print(df)
    print('----------')
    new_df = pd.concat([new_df, df]) # pd.concat() to concat two dataframes
print(new_df.shape)
```
